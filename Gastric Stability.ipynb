## DATA LOADING, CLEANING & ANALYSIS

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Yo/Stability/gastric_stability/gastric.csv", low_memory= False)
```

```python
df
```

```python
df = df[['TPSA', 'NumRotatableBonds', 'NumHDonors','NumHAcceptors', 'MolLogP', 'MolWt', 'Stability']]
```

```python
import warnings
from pandas.errors import SettingWithCopyWarning
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=SettingWithCopyWarning)
    df['Stability'] = label_encoder.fit_transform(df['Stability'])
```

```python
categories = label_encoder.classes_
encoding_map = dict(zip(categories, label_encoder.transform(categories)))

for category, integer in encoding_map.items():
    print(f"{category} -> {integer}")
```

```python
df.corr()
```

```python
X = df[['TPSA', 'NumRotatableBonds', 'NumHDonors','NumHAcceptors', 'MolLogP', 'MolWt']]
y = df[['Stability']]
```

```python
X
```

```python
print(y.value_counts())
y = np.array(y).ravel()
```

```python
X.describe()
```

## SCALING AND EVALUATION USING cross_val_score

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X))
```

```python
import pickle 
with open('scaler2.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import make_scorer, f1_score, roc_auc_score, accuracy_score
from sklearn.preprocessing import label_binarize
```

```python
svc = SVC(kernel='rbf', random_state=42)
```

```python
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
    
    accuracies = cross_val_score(svc, X, y, cv=5, scoring='accuracy')
    f1_scores = cross_val_score(svc, X, y, cv=5, scoring='f1_weighted')

    print(f'Accuracy: Mean = {np.mean(accuracies)}, Std = {np.std(accuracies)}')
    print(f'F1 Score: Mean = {np.mean(f1_scores)}, Std = {np.std(f1_scores)}')
```

```python
svc.fit(X,y)
```

```python
import pickle
with open('gastric_stability_model.pkl', 'wb') as f:
    pickle.dump(svc, f)
```
