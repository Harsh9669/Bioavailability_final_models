## DATA LOADING, CLEANING & ANALYSIS

```python
import pandas as pd
import numpy as np

df = pd.read_csv("D:/Yo/Stability/intestinal_stability/intestine.csv", low_memory= False)
```

```python
df
```

```python
df = df[['TPSA', 'NumRotatableBonds', 'NumHDonors','NumHAcceptors', 'MolLogP', 'MolWt', 'Stability']]
```

```python
import warnings
from pandas.errors import SettingWithCopyWarning
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=SettingWithCopyWarning)
    df['Stability'] = label_encoder.fit_transform(df['Stability'])
```

```python
categories = label_encoder.classes_
encoding_map = dict(zip(categories, label_encoder.transform(categories)))

for category, integer in encoding_map.items():
    print(f"{category} -> {integer}")
```

```python
df.corr()
```

```python
X = df[['TPSA', 'NumRotatableBonds', 'NumHDonors','NumHAcceptors', 'MolLogP', 'MolWt']]
y = df[['Stability']]
```

```python
X
```

```python
print(y.value_counts())
y = np.array(y).ravel()
```

```python
X.describe()
```

## SCALING AND EVALUATION USING cross_val_score

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X))
```

```python
# Check min and max values of the scaler
print("Scaler min:", scaler.data_min_)
print("Scaler max:", scaler.data_max_)
```

```python
from sklearn.model_selection import cross_val_score
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import make_scorer, f1_score, roc_auc_score, accuracy_score
from sklearn.preprocessing import label_binarize
```

```python
xgb = xgb.XGBClassifier(random_state=42)
```

```python
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
    
    accuracies = cross_val_score(xgb, X, y, cv=5, scoring='accuracy')
    f1_scores = cross_val_score(xgb, X, y, cv=5, scoring='f1_weighted')

    print(f'Accuracy: Mean = {np.mean(accuracies)}, Std = {np.std(accuracies)}')
    print(f'F1 Score: Mean = {np.mean(f1_scores)}, Std = {np.std(f1_scores)}')
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 300, 500],  
    'learning_rate': [0.01, 0.05, 0.1],  
    'max_depth': [3, 7, 9],  
    'min_child_weight': [1, 4, 7],  
    'gamma': [0, 0.1, 0.5], 
    'subsample': [0.7, 1.0],  
    'colsample_bytree': [0.7, 1.0] 
}
```

```python
xgb_clf = XGBClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=xgb_clf,
    param_grid=param_grid,
    scoring='f1_macro',  
    cv=5,  
    verbose=1,  
    n_jobs=-1 
)
```

```python
grid_search.fit(X, y)
```

```python
print("Best Parameters: ", grid_search.best_params_)
print("Best F1 Score: ", grid_search.best_score_)
```

```python
xgb_clf = XGBClassifier(
    colsample_bytree=1.0,
    gamma=0.1,
    learning_rate=0.1,
    max_depth=7,
    min_child_weight=1,
    n_estimators=500,
    subsample=0.7,
    random_state=42
)
```

```python
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
    
    accuracies = cross_val_score(xgb_clf, X, y, cv=5, scoring='accuracy')
    f1_scores = cross_val_score(xgb_clf, X, y, cv=5, scoring='f1_weighted')

    print(f'Accuracy: Mean = {np.mean(accuracies)}, Std = {np.std(accuracies)}')
    print(f'F1 Score: Mean = {np.mean(f1_scores)}, Std = {np.std(f1_scores)}')
```

```python
xgb_clf.fit(X,y)
```

```python
import pickle
with open('intestinal_stability_model.pkl', 'wb') as f:
    pickle.dump(xgb_clf, f)
```
